```python
import math
# 设置所需的精度（小数点后位数）
a1, b1, c1, a2, b2, c2 = 0, 0, 0, 0, 0, 0
lim = int(1e6)

def checkf(x, y): # x, y回代是否可以满足
    if x < -lim or x > lim or y < -lim or y > lim or x != int(x) or y != int(y):
        return False
    if a1 * x + b1 * y != c1:
        return False
    return a2 * (x ** 2) + b2 * (y ** 2) == c2

def Try(y):
    x = round((c1 - b1 * y) / a1)
    if checkf(x, y):
        print("Yes\n%d %d" % (x, y)); return True
    return False


def check(x, y):
    if (x * a1 + y * b1 == c1 and x ** 2 * a2 + y ** 2 * b2 == c2):
        if (-lim <= x <= lim and -lim <= y <= lim):
            return True
    else:
        return False

def asrt(x : int, y : int):
    if (check(x, y)):
        print("YES")
        print("%d %d" % (x, y))
        return True
    else:
        print("NO")
        return False

def exgcd(a : int, b : int):
    if (b == 0):
        return (1, 0, a)
    la, lb, r = exgcd(b, a % b)
    x = lb
    y = la - lb * (a // b)
    return x, y, r

def linear():
    x0, y0, r = exgcd(a1, b1)
    if (c1 % r != 0):
        print("NO")
        return
    x0 = (c1 // r) * x0
    y0 = (c1 // r) * y0
    k = b1 // r; l = - a1 // r
    # -lim <= i * k + x0 <= lim
    # -lim <= - i * l + y0 <= lim
    x_i_up = (lim - x0) / k # x_i * k <= lim - x0
    x_i_low = (-lim - x0) / k # x_i * k
    x_i_up, x_i_low = (x_i_low, x_i_up) if x_i_up < x_i_low else (x_i_up, x_i_low)

    y_i_up = (lim - y0) / l # x_i * k <= lim - x0
    y_i_low = (-lim - y0) / l # x_i * k
    y_i_up, y_i_low = (y_i_low, y_i_up) if y_i_up < y_i_low else (y_i_up, y_i_low)

    i_low = int(max(y_i_low, x_i_low))
    i_up = int(min(y_i_up, x_i_up))

    if (i_low > i_up):
        print("NO")
        return
    
    x = i_low * k + x0
    y = i_low * l + y0
    asrt(x, y)
    return


def solve():
    if (a1 == 0 and b1 == 0):
        print("NO")
        return
    
    if (a1 == 0): # b1 != 0
        y = c1 // b1
        if (y * b1 != c1):
            print("NO")
            return
        else:
            if (a2 != 0):
                if (((c2 - b2 * y ** 2) // a2) < 0):
                    print("NO")
                    return
                x = round(math.sqrt((c2 - b2 * y ** 2) // a2))
            else:
                x = 0
            asrt(x, y)
            return

    if (b1 == 0):
        x = c1 // a1
        if (x * a1 != c1):
            print("NO")
            return
        else:
            if (b2 != 0):
                if (((c2 - a2 * x ** 2) // b2) < 0):
                    print("NO")
                    return
                y = round(math.sqrt((c2 - a2 * x ** 2) // b2))
            else:
                y = 0
            asrt(x, y)
            return
    
    if (a2 == 0 and b2 == 0 and c2 != 0):
        print("NO")
        return
    if (a2 == 0 and b2 == 0 and c2 == 0):
        linear()
        return
    
    if (a2 == 0 or b2 == 0):
        if (a2 == 0):
            if (c2 // b2 < 0):
                print("NO")
                return
            y1 = int(math.sqrt(c2 // b2))
            x1 = (c1 - y1 * b1) // a1
            if(check(x1, y1)):
                asrt(x1, y1)
                return
            y2 = -y1
            x2 = (c1 - y2 * b1) // a1
            asrt(x2, y2)
            return
        
        if (b2 == 0):
            if (c2 // a2 < 0):
                print("NO")
                return
            x1 = int(math.sqrt(c2 // a2))
            y1 = (c1 - x1 * a1) // b1
            if(check(x1, y1)):
                asrt(x1, y1)
                return
            x2 = -x1
            y2 = (c1 - x2 * a1) // b1
            asrt(x2, y2)
            return
    
    # a1 x + b1 x = c1
    # a2 x**2 + b2 y**2 = c2

    # A = int(a2 * b1 ** 2 + a1**2 * b2)
    # B = int(-2 * a2 * b1 * c1)
    # C = int(a2 * c1 ** 2 - a1 ** 2 * c2)
    # # common_gcd = math.gcd(A, B)
    # # common_gcd = math.gcd(common_gcd, C)
    # # A = A // common_gcd
    # # B = B //common_gcd
    # # C = C // common_gcd
    # num = Decimal(B ** 2 - 4 * A * C)
    # delta = int(num.sqrt())
    # if (delta ** 2 != B ** 2 - 4 * A * C):
    #     print("NO")
    #     return
    # if ((-B + delta) % (2 * A) == 0):
    #     result_1 = (-B + delta) // (2 * A)
    #     x = (c1 - b1 * result_1) // a1
    #     asrt(x, result_1)
    #     return
    # if ((-B - delta) % (2 * A) == 0):
    #     result_2 = (-B + delta) // (2 * A)
    #     x = (c1 - b1 * result_2) // a1
    #     asrt(x, result_2)
    #     return
    A = (a2 * b1 * b1) / (a1 * a1) + b2
    B = -(2 * a2 * c1 * b1) / (a1 * a1)
    C = (a2 * c1 * c1) / (a1 * a1) - c2
    Delta = round(B * B - 4 * A * C, 4)
    if A == 0:
        if B == 0:
            y1 = y2 = 0
        else:
            y1 = y2 = round((-C) / B)
    else:
        if Delta < 0:
            print("No", Delta); return
        BA = (-B) / (2 * A)
        y1 = round(BA - math.sqrt(Delta / (4 * A * A)))
        y2 = round(BA + math.sqrt(Delta / (4 * A * A)))
    flag = True
    for d in range(-5, 5):
        if flag:
            y = y1 + d
            if Try(y):
                flag = False; break
            y = y2 + d
            if Try(y):
                flag = False; break
    if flag:
        print("No")
```

```python
T = int(input())
for _ in range(T):
    a1, b1, c1, a2, b2, c2 = map(int, input().split()) # 接收a, b, c
    solve()
```
